```python
import numpy as np
import matplotlib.pyplot as plt

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot with rainbow colormap
plt.scatter(x, y, c=y, cmap='rainbow')

plt.colorbar()  # show color scale
plt.title("Rainbow Colormap Example")
plt.show()
```

```python
x= np.asarray(x)
y= np.asarray(y)
x_midpts = np.hstack((x[0], 0.5 * (x[0] + x[1]), x[1:]))
y_midpts = np.hstack((y[0], 0.5 * (y[0] + y[1]), y[1:]))
print(x_midpts)
print(y_midpts)
```

```python
category = ("finance", "sales", "hr", "research", "technology")
```

```python
expenditure = np.random.randint(30000, 200000, 5)
```

```python
plt.bar(category, expenditure ,width=0.9, color=["teal","green","olivedrab","darkmagenta","orange"])
plt.xlabel("category")
plt.ylabel("Expenditure")
plt.title("Expenditure by Category")
plt.show()
```

```python
plt.barh(category, expenditure,height=0.3, color=["teal","green","olivedrab","darkmagenta","orange"])
plt.xlabel("Expenditure")
plt.ylabel("Category")
plt.title("Expenditure by Category")
plt.show()
```

```python
data= np.random.randn(1000)
data=np.random.choice([20,40, 250, 400, 150,350], 1000)
data=np.random.randint(20, 400, 1000)
plt.figure(figsize=(16,9))  
plt.title("Histogram of Random Data")
plt.hist(data, bins=20, color='steelblue', rwidth=0.8)  
plt.xlabel("Value")
plt.ylabel("Frequency") 
plt.show()
```

```python
plt.pie(expenditure, labels=category, autopct='%0.0f%%', hatch=['**0', 'o0', '0.0', '||', 'x'])
plt.title("Pie Chart Example")
plt.show()
```

```python
plt.pie(expenditure, labels=category, autopct='%0.0f%%', pctdistance=1.23,labeldistance=0.3)
plt.title("Pie Chart Example")
plt.show()
```

```python
explode=(0,0.1,0,0,0) 
plt.pie(expenditure, labels=category, autopct='%0.0f%%', pctdistance=1.23,labeldistance=0.3,explode=explode,shadow=True,startangle=180)
plt.title("Pie Chart Example")
plt.show()
```
